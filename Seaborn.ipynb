```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
sns.get_dataset_names()
```

```python
df=sns.load_dataset('tips')
```

```python
df
```

```python
#relplot gives the relationship between two categorical or numerical variables
sns.relplot(x='total_bill',y='tip',data=df)
```

```python
sns.relplot(x='total_bill',y='tip',data=df,kind='line')
```

```python
sns.relplot(x='total_bill',y='tip',data=df,hue='smoker') #hue gives different color coding
```

```python
sns.relplot(x='total_bill',y='tip',data=df,style='size') #style gives different representations of column
```

```python
sns.relplot(x='total_bill',y='tip',data=df,hue='smoker',style='size')
```

```python
time=np.arange(500)
value=np.random.randn(500).cumsum()
```

```python
df=pd.DataFrame({'time':time,'value':value})
df
```

```python
sns.relplot(x='time',y='value',kind='line',data=df,sort=True)
```
