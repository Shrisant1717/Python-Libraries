```python
x1=[1,2,3,4,5]
y1=[5,1,7,8,2]
y2=[7,8,3,5,1]
plt.plot(x1,y1,color='yellow',marker='o',linestyle='--',linewidth=3,markersize=6,label='line y1')
plt.plot(x1,y2,color='red',marker='o',linestyle='--',linewidth=3,markersize=6,label='line y2')
plt.legend()
plt.grid()
plt.show()
```

```python
x=np.linspace(1,10,100)
plt.plot(x,np.sin(x))
plt.plot(x,np.cos(x))
plt.show()
```

```python
x=np.linspace(1,10,100)
plt.plot(x,np.sin(x),'--')
plt.plot(x,np.cos(x))
plt.show()
```

```python
plt.figure()
plt.subplot(1,2,1)
plt.plot(x,np.sin(x))
plt.subplot(1,2,2)
plt.plot(x,np.cos(x))
plt.show()
```

```python
plt.figure()
plt.subplot(2,1,1)
plt.plot(x,np.sin(x))
plt.subplot(2,1,2)
plt.plot(x,np.cos(x))
plt.show()
```

```python
plt.figure()
plt.subplot(2,2,1)
plt.plot(x,np.sin(x))
plt.subplot(2,2,2)
plt.plot(x,np.cos(x))
plt.subplot(2,2,3)
plt.plot(x,np.tan(x))
plt.subplot(2,2,4)
plt.plot(x,np.sin(x))
plt.show()
```

```python
# Histogram
data=np.random.randn(1000)
plt.hist(data,edgecolor='white')
plt.show()
```

```python
# Histogram
data=np.random.randn(1000)
plt.hist(data,color='red',bins=20,edgecolor='white')
plt.show()
```

```python
# Staged histogram
data1=np.random.normal(0,1,5000)
data2=np.random.normal(3,2,5000)
```

```python
data1
```

```python
data2
```

```python
plt.hist([data1,data2],color=['r','b'],label=['dist1','dist2'],edgecolor='w')
plt.xlabel("Values")
plt.ylabel("Frequency")
plt.title("Staged Histogram")
plt.legend()
plt.show()
```

```python
# bar plot
# To see distribution of categorical data
x=['a','b','c','d']
y=[4,6,1,10]
plt.bar(x,y,color='red')
```

```python
# pie chart
langs=['C','Python','Java']
students=[20,100,40]
plt.pie(students,labels=langs,autopct="%1.1f%%")
plt.title("Pie Chart")
plt.show()
```

```python
# focus on one segment
langs=['C','Python','Java']
students=[20,100,40]
plt.pie(students,labels=langs,autopct="%1.1f%%",explode=(0.0,0.1,0.1),shadow="True")
plt.title("Pie Chart")
plt.show()
```

```python
# 3D plots
x=np.random.rand(20)
y=np.random.rand(20)
z=np.random.rand(20)
fig=plt.figure()
ax=fig.add_subplot(projection="3d")
plt.scatter(x,y,z,c='r',marker='o')
```
